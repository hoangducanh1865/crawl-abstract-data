```python
import fitz  # PyMuPDF
import os
import re

# ÄÆ°á»ng dáº«n tá»›i thÆ° má»¥c chá»©a file PDF
input_folder = "/Users/hoangducanh/Documents/Hoc o HUST/nhom_anh_minh/crawl-abstract-data/data/input/khcn/"  # Thay Ä‘á»•i Ä‘Æ°á»ng dáº«n
output_folder = "/Users/hoangducanh/Documents/Hoc o HUST/nhom_anh_minh/crawl-abstract-data/data/output/khcn/"  # ThÆ° má»¥c lÆ°u file text

def get_title(page):
    text_blocks = page.get_text("dict")["blocks"]

    # ğŸ“Œ Biáº¿n lÆ°u text cÃ³ cá»¡ chá»¯ lá»›n nháº¥t
    max_font_size = 0
    max_text = ""

    # ğŸ“Œ Duyá»‡t qua tá»«ng khá»‘i vÄƒn báº£n Ä‘á»ƒ tÃ¬m text cÃ³ cá»¡ chá»¯ lá»›n nháº¥t
    for block in text_blocks:
        if "lines" in block:  # Kiá»ƒm tra náº¿u cÃ³ dÃ²ng chá»¯
            for line in block["lines"]:
                for span in line["spans"]:  # "spans" chá»©a thÃ´ng tin vá» font vÃ  kÃ­ch thÆ°á»›c chá»¯
                    text = span["text"].strip()
                    font_size = span["size"]

                    if font_size > max_font_size:  # Cáº­p nháº­t náº¿u cá»¡ chá»¯ lá»›n hÆ¡n
                        max_font_size = font_size
                        max_text = text
                    elif font_size == max_font_size:
                        max_text += " " + text
                    elif max_font_size - font_size < 5 and text != "":
                        max_text += " " + text
                    
    # Giá»›i háº¡n Ä‘á»™ dÃ i cá»§a tÃªn tá»‡p
    max_text = max_text[:100]  # Giá»›i háº¡n tÃªn tá»‡p tá»‘i Ä‘a 100 kÃ½ tá»±
    return re.sub(r'[<>:"/\\|?*]', '_', max_text) #thay tháº¿ cÃ¡c kÃ­ tá»± Ä‘áº·c biá»‡t khÃ´ng thá»ƒ lÆ°u thÃ nh tÃªn file

# Duyá»‡t qua táº¥t cáº£ cÃ¡c file PDF trong thÆ° má»¥c
for filename in os.listdir(input_folder):
    if filename.endswith(".pdf"):  # Chá»‰ xá»­ lÃ½ file PDF
        pdf_path = os.path.join(input_folder, filename)
        
        # Má»Ÿ file PDF
        doc = fitz.open(pdf_path)
        page = doc[0]  # Chá»‰ xÃ©t trang Ä‘áº§u tiÃªn

        # Láº¥y danh sÃ¡ch Ä‘Æ°á»ng káº» ngang
        horizontal_lines = [
            l for l in page.get_drawings()
            if abs(l["rect"][1] - l["rect"][3]) < 2 and l["type"] == 's' # ÄÆ°á»ng káº» ngang
        ]

        if len(horizontal_lines) < 2:
            print("KhÃ´ng tÃ¬m tháº¥y Ä‘á»§ 2 Ä‘Æ°á»ng káº» ngang trÃªn trang Ä‘áº§u tiÃªn!")
            continue
    
        # Sáº¯p xáº¿p Ä‘Æ°á»ng káº» theo vá»‹ trÃ­ y (tá»« trÃªn xuá»‘ng dÆ°á»›i)
        horizontal_lines.sort(key=lambda l: l["rect"][1])

        # XÃ¡c Ä‘á»‹nh vá»‹ trÃ­ cá»§a 2 Ä‘Æ°á»ng káº» ngang Ä‘áº§u tiÃªn
        if len(horizontal_lines) == 2:
            x1 = horizontal_lines[0]["rect"][0]  # Vá»‹ trÃ­ x bÃªn trÃ¡i cá»§a Ä‘Æ°á»ng Ä‘áº§u tiÃªn
            y1 = horizontal_lines[0]["rect"][1]  # Vá»‹ trÃ­ y cá»§a Ä‘Æ°á»ng Ä‘áº§u tiÃªn
            x2 = horizontal_lines[1]["rect"][2]  # Vá»‹ trÃ­ x bÃªn pháº£i cá»§a Ä‘Æ°á»ng thá»© hai
            y2 = horizontal_lines[1]["rect"][1]  # Vá»‹ trÃ­ y cá»§a Ä‘Æ°á»ng thá»© hai
        else:
            x1 = horizontal_lines[1]["rect"][0]
            y1 = horizontal_lines[1]["rect"][1]  # Vá»‹ trÃ­ y cá»§a Ä‘Æ°á»ng Ä‘áº§u tiÃªn
            x2 = horizontal_lines[2]["rect"][2]
            y2 = horizontal_lines[2]["rect"][1]  # Vá»‹ trÃ­ y cá»§a Ä‘Æ°á»ng thá»© hai
   
        # Láº¥y cÃ¡c khá»‘i vÄƒn báº£n trong khoáº£ng y1 < text < y2
        text_blocks = page.get_text("blocks")
        extracted_text = [
            block[4] for block in text_blocks if y1 < block[1] < y2 and block[2] >= x1 and block[0] <= x2
        ]

        title = get_title(page)  # Láº¥y tiÃªu Ä‘á» Ä‘áº§u tiÃªn
        output_txt = os.path.join(output_folder, f"{title}.txt")

        # Ghi káº¿t quáº£ vÃ o file text
        try:
            with open(output_txt, "w", encoding="utf-8") as f:
                f.write("\n".join(extracted_text))
                print('âœ…')
        except OSError as e:
            print(f"Error writing file {output_txt}: {e}")

        doc.close()

print("Xá»­ lÃ½ xong táº¥t cáº£ file PDF!")
```
